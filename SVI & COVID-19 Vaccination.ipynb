```python
import pandas as pd
```

```python
data = pd.read_csv('covidvaccinations.csv')
```

```python
df = pd.DataFrame(data)
```

```python
df
```

```python
INdata = df[df['Recip_State']=='IN']
```

```python
INdata
```

```python
INdata['Recip_County'].nunique()
```

```python
INdata = INdata[INdata['Date']=='10/15/2021']
```

```python
INdata
```

```python
county = []
for i in INdata['Recip_County']:
    county.append(i)
```

```python
print(county)
```

```python
INdata = INdata[INdata.Recip_County != 'Unknown County']
```

```python
len(county)
```

```python
INdata
```

```python
INdata = INdata[['Recip_County','Series_Complete_Pop_Pct', 'FIPS']]
```

```python
INdata = INdata.sort_values(by=['Recip_County'])
```

```python
INdata.dtypes
```

```python
INdata
```

```python
INdata["FIPS"] = INdata["FIPS"].astype(object).astype(int)
```

```python
INdata.dtypes
```

```python
df2 = pd.DataFrame(data2)
```

```python
df2
```

```python
df2 = df2[['COUNTY', 'RPL_THEME1', 'RPL_THEME2', 'RPL_THEME3', 'RPL_THEME4', 'RPL_THEMES', 'FIPS']]
```

```python
df2
```
